```python

import random

def number():
    while True:
        a = random.randint(4,10)
        b = random.randint(4,10)
        c = random.randint(4,10)
        d = 25 - a-b-c
        if d <=0:
            pass

        elif (d<=10)and(d>=4):
            break
        else:
            pass
            
    return a, b,c,d
```
